Hacemos una función que devuelva en un diccionario todos los datos de las películas, salvo la fecha de estreno y la url

Parámetros: url y fecha de estreno
Salida: Diccionario con todos los datos

```python
def get_datos_pelicula(driver, url, fecha):
    driver.get(url)
    
    wait = WebDriverWait(driver, 10)
    
    pelicula = {
        "fecha_estreno": fecha,
        "url": url
    }
    
    wait.until(EC.presence_of_element_located((By.XPATH, '//*[contains(@class, "mtitle")]')))
    
    try:
        titulo = driver.find_element(By.XPATH, '//*[contains(@class, "mtitle")]').text
        pelicula["titulo"] = titulo
    except:
        pelicula["titulo"] = None

    try:
        nota = driver.find_element(By.XPATH, '//*[contains(@itemprop,"ratingValue")]').text
        pelicula["nota_media"] = nota
    except:
        pelicula["nota_media"] = None

    try:
        votos = driver.find_element(By.XPATH, '//*[contains(@itemprop,"ratingCount")]').text
        pelicula["cantidad_votos"] = votos
    except:
        pelicula["cantidad_votos"] = None

    try:
        labels = driver.find_elements(By.XPATH, '//dl[contains(@class,"movie-info")]//dt')
        values = driver.find_elements(By.XPATH, '//dl[contains(@class,"movie-info")]//dd')
        
        for lab, val in zip(labels, values):
            clave = lab.text.strip().replace(":", "")
            valor = val.text.strip()
            pelicula[clave] = valor
    except:
        pass
    
    return pelicula
```

Sacamos todas las películas y llamamos a la función con cada película

```python
peliculas_cartelera = driver.find_elements(By.XPATH, '//div[@class="col"]')
print(peliculas_cartelera)
```

```python
links = []
for pelicula in peliculas_cartelera:
    enlace = pelicula.find_element(By.TAG_NAME, "a")
    url = enlace.get_attribute("href")
    titulo = enlace.get_attribute("title")
    links.append({
        "titulo": titulo,
        "url": url
    })
```

Ahora usamos los links para llamar a la funcion y sacar los datos

```python
datos_de_peliscartelera = []

for peliculas in links:
    datos = get_datos_pelicula(driver, peliculas["url"], "fecha")
    datos_de_peliscartelera.append(datos)
```
